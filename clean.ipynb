```python
import pandas as pd
```

```python
def replace_species(df, to_replace, replace_with):
    
    df.loc[df['species'] == to_replace, 'species'] = replace_with
    
    return(df)
```

```python
df = pd.read_csv('AWI_barnfires.csv')
```

```python
df = df.drop(df[df.species == 'species'].index).reset_index(drop=True)
```

```python
df = df.sort_values(by = 'species').reset_index(drop=True)
```

```python
df
```

```python
replacements = {
    'Animals (species unspecified)':'Animals (species not specified)',
    'Calf':'Calves',
    'Cow':'Cows',
    'Horse':'Horses'
}
```

```python
df = df.replace(replacements)
```

```python
# Get all unique species 
df['species'].unique()
```

```python
# Separate out muliple species - break dataset into 2 
m_df = df[df['species'].str.contains('Multiple species')].reset_index(drop = True)
```
